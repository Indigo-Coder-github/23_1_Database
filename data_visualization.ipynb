<h1>범주형 데이터 확인</h1>

```python
cate_data = pd.DataFrame()
for col in data:
    if data[col].dtype == object or data[col].nunique() <= 10:
        cate_data[col] = data[col]
        print(f'{col} : {data[col].unique()}')
        print(data[col].value_counts())
        print()
```

```python
cate_data.info()
```

```python
change_dict = {1: 'Below College', 2: 'College', 3: 'Bachelor', 4: 'Master', 5: 'Doctor'}
cate_data.replace({'Education': change_dict}, inplace=True)

change_dict = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}
cate_data.replace({'EnvironmentSatisfaction': change_dict}, inplace=True)

change_dict = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}
cate_data.replace({'JobInvolvement': change_dict}, inplace=True)

change_dict = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}
cate_data.replace({'JobSatisfaction': change_dict}, inplace=True)

change_dict = {1: 'Low', 2: 'Good', 3: 'Excellent', 4: 'Outstanding'}
cate_data.replace({'PerformanceRating': change_dict}, inplace=True)

change_dict = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}
cate_data.replace({'RelationshipSatisfaction': change_dict}, inplace=True)

change_dict = {1: 'Bad', 2: 'Good', 3: 'Better', 4: 'Best'}
cate_data.replace({'WorkLifeBalance': change_dict}, inplace=True)
```

```python
fig, ax = plt.subplots(4, 5, figsize=(30, 30))
ax = ax.flatten()

for i in range(len(cate_data.columns)):
    sns.countplot(x=cate_data.columns[i], data=cate_data, ax=ax[i], hue = 'Attrition').set(xlabel=None, ylabel=None)
    ax[i].set(title=cate_data.columns[i])
    
    if len(str(cate_data[cate_data.columns[i]].unique().max())) > 10:
        ax[i].tick_params(axis='x', labelrotation=20)
```

```python
fig = plt.figure(figsize=(15, 15))

sns.countplot(x = data['NumCompaniesWorked'], hue='Attrition', data=data)
```

```python
fig, ax = plt.subplots(4, 5, figsize=(30, 30))
ax = ax.flatten()

for i in range(len(cate_data.columns)):
    col = cate_data.columns[i]
    count_data = cate_data.groupby([col, 'Attrition']).size().unstack().reset_index()
    count_data['Attrition_Rate'] = count_data['Yes'] / (count_data['Yes'] + count_data['No'])
    sns.barplot(x=col, y='Attrition_Rate', data=count_data, ax = ax[i])
    
    if len(str(cate_data[col].unique().max())) > 10:
        ax[i].tick_params(axis='x', labelrotation=20)
```

```python
fig, ax = plt.subplots(4, 5, figsize=(30, 30))
ax = ax.flatten()

for i in range(len(cate_data.columns)):
    col = cate_data.columns[i]
    count_data = cate_data.groupby([col, 'Attrition']).size().unstack().reset_index()
    count_data['Attrition_Rate'] = count_data['No'] / (count_data['Yes'] + count_data['No'])
    sns.barplot(x=col, y='Attrition_Rate', data=count_data, ax = ax[i])
    
    if len(str(cate_data[col].unique().max())) > 10:
        ax[i].tick_params(axis='x', labelrotation=20)
```

<h1>연속형 데이터 확인</h1>

```python
cont_data = pd.DataFrame()
for col in data:
    if data[col].dtype != object and data[col].nunique() > 10:
        cont_data[col] = data[col]
        print(f'{col} : {data[col].nunique()}')
```

```python
print(data['YearsAtCompany'].unique())
```

```python
fig, ax = plt.subplots(4, 4, figsize=(30, 30))
ax = ax.flatten()
cont_data['Attrition'] = data['Attrition']

for i in range(len(cont_data.columns)):
    if cont_data.columns[i] != 'Attrition':
        sns.distplot(cont_data[cont_data['Attrition']=='Yes'][cont_data.columns[i]],kde = True, color='r',ax=ax[i], hist = False, bins=20,label = 'Yes')
        sns.distplot(cont_data[cont_data['Attrition']=='No'][cont_data.columns[i]], kde = True, color='b',ax=ax[i], hist = False, bins=20, label='No')
        ax[i].legend()
```

```python
fig, ax = plt.subplots(4, 4, figsize=(30, 30))
ax = ax.flatten()
cont_data['Attrition'] = data['Attrition']

for i in range(len(cont_data.columns)):
    if cont_data.columns[i] != 'Attrition':
            if cont_data.columns[i] in ['Age', 'DailyRate', 'DistanceFromHome', 'EmployeeNumber', 'MonthlyIncome', 'YearsAtCompany', 'YearsWithCurrManager','YearsInCurrentRole']:
                sns.distplot(cont_data[cont_data['Attrition']=='Yes'][cont_data.columns[i]],kde = True, color='r',ax=ax[i], hist = False, bins=20,label = 'Yes')
                sns.distplot(cont_data[cont_data['Attrition']=='No'][cont_data.columns[i]], kde = True, color='b',ax=ax[i], hist = False, bins=20, label='No')
                ax[i].legend()
```

```python
fig, ax = plt.subplots(4, 4, figsize=(30, 30))
ax = ax.flatten()

for i in range(len(cont_data.columns)):
    if cont_data.columns[i] != 'Attrition':
        sns.boxplot(x='Attrition', y=cont_data.columns[i], data=cont_data, ax=ax[i])
```
